# Demo: EMG Analysis Pipeline
This notebook demonstrates the core functionality of the EMG Analysis Engine.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from core_engine import EMGConfig, EMGFeatureExtractor, EMGSignalSimulator

# Load configuration
config = EMGConfig()
engine = EMGFeatureExtractor(config)

# Generate synthetic signal
sim = EMGSignalSimulator()
signal = sim.generate_contraction(duration=3.0, sampling_rate=2000, intensity=1.0)

# Preprocess
filtered = engine.preprocess(signal)

# Extract features
features = engine.process_stream(signal)

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(2,1,1)
plt.plot(signal, alpha=0.5, label='Raw')
plt.plot(filtered, label='Filtered')
plt.title('Raw vs Filtered EMG')
plt.legend()

plt.subplot(2,1,2)
rms = [f['RMS'] for f in features['time_series']['features']]
plt.plot(rms, label='RMS Envelope')
plt.title('RMS Feature over Time')
plt.legend()
plt.show()

print("✅ Analysis complete. Summary:", features['summary_statistics'])
```

## Results
The above plots show the raw EMG signal, the filtered signal, and the RMS envelope over time.
